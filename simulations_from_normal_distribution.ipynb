# Simulations: Normal Distribution

```python
# Codes to generate rendom numbers from loc and scaloc mediated normal distributions.

def loc_simulation(mu, sigma, sample_size, iteration):
    random.seed(1234)
    multiple_list = []
    for seed in np.random.randn(iteration):
        random.seed(seed)
        multiple_list.append(np.random.randn(sample_size)*sigma+mu)
    simulation = [k for k in (multiple_list)]
    return (simulation)

def scaloc_simulation(mu, sigma, sample_size, iteration):
    random.seed(1234)
    multiple_list = []
    for seed in np.random.randn(iteration):
        random.seed(seed)
        multiple_list.append((np.random.randn(sample_size)+mu)*sigma)
    simulation = [k for k in (multiple_list)]
    return (simulation)

def loc_mixsimulation(mu, sigma, sample_size, iteration):
    random.seed(1234)
    multiple_list = []
    for seed in np.random.randn(iteration):
        random.seed(seed)
        mu2 = [mu, 15]
        sigma2 = [sigma, 10]
        p_i = [0.9, 0.1]
        #sample_size = 1000
        x = [ ]
        for i in range(sample_size):
            z_i = np.argmax(np.random.multinomial(1, p_i))
            x_i = np.random.normal(mu2[z_i], sigma2[z_i])
            x.append(x_i)
        multiple_list.append(x)
    simulation = [k for k in (multiple_list)]
    return (simulation)

def scaloc_mixsimulation(mu, sigma, sample_size, iteration):
    random.seed(1234)
    multiple_list = []
    for seed in np.random.randn(iteration):
        random.seed(seed)
        mu2 = [mu, 10]
        sigma2 = [sigma, 10]
        p_i = [0.8, 0.2]
        #sample_size = 1000
        x = [ ]
        for i in range(sample_size):
            z_i = np.argmax(np.random.multinomial(1, p_i))
            x_i = np.random.normal(sigma2[z_i], mu2[z_i])
            x.append(x_i)
        multiple_list.append(x)
    simulation = [k for k in (multiple_list)]
    return (simulation)
```
